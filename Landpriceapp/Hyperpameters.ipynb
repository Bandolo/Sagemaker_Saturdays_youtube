```python
#Importing the necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
trainset = pd.read_csv('train_cleaned.csv')
trainset
```

```python
testset = pd.read_csv('test_cleaned.csv')
testset
```

```python
X = trainset.drop('Price_log',axis = 1)
X
```

```python
Y = trainset['Price_log']
```

```python
# Hyper Parameter Optimization
params ={
    "learning_rate"       : [.05, 0.10, 0.15, 0.20, 0.25, 0.30 ],
    "max_depth"           : [3, 4, 5, 6, 8, 10, 12, 15],
    "min_child_weight"    : [1, 3, 5, 7 ],
    "gamma"               : [0.0, 0.1, 0.2, 0.3, 0.4],
    "colsample_bytree"    : [0.3, 0.4, 0.5, 0.7]
    
}
```

```python
#Hyperparameter optimzation using Randomized SearchCV
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
#from numpy import nan
```

```python
regressor = xgboost.XGBRegressor()
```

```python
random_search = RandomizedSearchCV(regressor,param_distributions=params,n_iter=5,scoring='neg_mean_squared_error', n_jobs=-1,cv=5,verbose=3)
```

```python
random_search.fit(X,Y)
```

```python
random_search.best_estimator_
```
